Titanic dataset is one of the datasets available in sklearn. You are given:

1. A Training dataset csv file with X train and Y train data
2. A X test File and you have to predict and submit predictions for this file.


Your task is to:

- Use Logistic Regression and come with predictions.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#replace null value with mean value
df.Age.fillna(df.Age.mean(),inplace=True)
df.describe()
```

```python
def f1(s):
    if s=='female':
        return 0
    else:
        return 1

def f2(s):
    if s=='S':
        return 1
    elif s=='Q':
        return 2
    else:
        return 3
def f3(s):
    return s
```

```python
df["sex_numeric"]=df.Sex.apply(f1)
df["Embarked_numeric"]=df.Embarked.apply(f2)
df["Survived_numeric"]=df.Survived.apply(f3)
del df["Name"]
del df["Cabin"]
del df["Sex"]
del df["Embarked"]
del df["Ticket"]
del df["Fare"]
del df["Survived"]
df.head()
```

```python
df.dtypes
```

```python
train_data=np.array(df,int)
train_data
```

```python
x_train=train_data[:,:-1]
y_train=train_data[:,-1]
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
#train the classifier
clf.fit(x_train,y_train)
```

```python
df2=pd.read_csv("0000000000002429_test_titanic_x_test.csv")
print(df2.shape)
df2.describe()
```

```python
df2.head()
```

```python
df2.Age.fillna(df2.Age.mean(),inplace=True)
df2["sex_numeric"]=df2.Sex.apply(f1)
df2["Embarked_numeric"]=df2.Embarked.apply(f2)
del df2["Name"]
del df2["Sex"]
del df2["Embarked"]
del df2["Cabin"]
del df2["Ticket"]
del df2["Fare"]
df2.head()
```

```python
test_data=np.array(df2,int)
test_data
```

```python
x_test=test_data[:,:]
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
y_pred=y_pred.reshape(-1,1)
```

```python
np.savetxt("output_logistic_titanic.csv",y_pred)
```
